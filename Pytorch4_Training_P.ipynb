# **Defining Model**
* Defining mean desigining structure of the neural network

```python
class MyNN():

  def __init__(self,X):

    self.weights = torch.rand(X.shape[1],1,requires_grad=True,dtype=torch.float64)
    self.bias = torch.zeros(1,requires_grad=True,dtype=torch.float64)

  def forward(self,X):
    X = X.to(torch.float64)
    z = torch.matmul(X,self.weights) + self.bias

    y_pred = torch.sigmoid(z)
    return y_pred

  def loss_function(self,y_pred,y):
    # clamp predictions to avoid log(0)
    epsilon =1e-7
    y_pred = torch.clamp(y_pred,epsilon,1-epsilon)


    #calculate  loss
    loss = -(y * torch.log(y_pred)+ (1- y) * torch.log(1 - y_pred)).mean()

    return loss





```

# **Important Parameter**

```python
learning_rate = 0.1
epochs = 25
```

# **Training PIPELINE**

```python
# step 1- first we create our model
model = MyNN(x_train_tensor)

# step 2 - Training model

 # define loop for epochs
for epoch in range(epochs):
  #(i): forward pass
  y_pred = model.forward(x_train_tensor)


  #(ii): loss calculate
  loss = model.loss_function(y_pred,y_train_tensor)


  #(iii): Backward pass (here we calculate the derivatives )
  loss.backward()

  #(iv): parameters update
  with torch.no_grad():
    model.weights -= learning_rate * model.weights.grad
    model.bias -= learning_rate * model.bias.grad

  # zero gradients
  model.weights.grad.zero_()
  model.bias.grad.zero_()

  # print loss in each epochs
  print(f'Epoch: {epoch+1}, Loss : {loss.item()}')
```

# **Model Desigining**

```python
class NN():

  def __init__(self,X):

    self.weights = torch.rand(X.shape[1],1,requires_grad=True,dtype=torch.float64)
    self.bias = torch.zeros(1,requires_grad=True,dtype=torch.float64)

  def forward(self,X):
    z = torch.matmul(X,self.weights) + self.bias

    y_pred = torch.sigmoid(z)
    return y_pred

  def loss_function(self,y_pred,y):
    # clamp predictions to avoid log(0)
    epsilon =1e-7
    y_pred = torch.clamp(y_pred,epsilon,1-epsilon)

    #calculate  loss
    loss = -(y * torch.log(y_pred)+ (1- y) * torch.log(1 - y_pred)).mean()
    return loss




```

# **Parameters**

```python
learning_rate = 0.1
epochs = 100
```

# **Training pipeline**

```python
# model creation

model = NN(x_train_tensor)

# loop for epochs
for epoch in range(epochs):
  # forward pass
  y_pred = model.forward(x_train_tensor)


  # loss function
  loss = model.loss_function(y_pred,y_train_tensor)

  # Backward propagation
  loss.backward()

  # update parameters
  with torch.no_grad():
    model.weights -= learning_rate * model.weights.grad
    model.bias -= learning_rate * model.bias.grad

  # set gradient to zero for next iteration
  model.weights.grad.zero_()
  model.bias.grad.zero_()

  # printing the result
  print(f'epochs: {epoch+1} Loss: {loss}')




```
